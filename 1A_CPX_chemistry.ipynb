```r
df <- read.table("CPX_experiments.txt",sep="\t",header=TRUE)
```

The line of code above reads in the table experimental data for the cpx.

Below we can plot this data, to understand how the chemistry of the cpx varies with pressure. You can do this by changing the x-variable.

We can also change the labels of the graph, the point type, size and colour. More details can be found here: https://www.statmethods.net/advgraphs/parameters.html

```r
yvar <- df$P_GPa
xvar <- df$FeOt.cpx
ylab <- "Pressure (Gpa)"
xlab <- "FeO (tot) (wt. %)"
symbol <- 12
symbol.size <- 1
col.points <- "pink"

plot(xvar,yvar,pch=symbol,col=col.points,cex=symbol.size,xlab=xlab,ylab=ylab) #do not change
```

We can also examine the relationships between pressure and cpx chemistry, by examining the effect of temperature or melt composition on cpx chemistry.

We do this by colour-contouring the points in the graph below. You can change the variable, number of colours and limits of the colour range. You can also change the position of the legend. As before, you can change other p

```r
yvar <- df$P_GPa
xvar <- df$SiO2.cpx
ylab <- "Pressure (Gpa)"
xlab <- "SiO2 (wt. %)"
symbol <- 16
symbol.size <- 1

colourvar <- df$T_K
colourvarname <- "Temperature (K)"
n.colours <- 5
pal <- colorRampPalette(c("red", "yellow"))
legend.position <- "topleft"

col.points <- pal(n.colours)[cut(colourvar,breaks = n.colours)] #do not change
plot(xvar,yvar,pch=symbol,col=col.points,cex=symbol.size,xlab=xlab,ylab=ylab) #do not change
legend(legend.position,title = colourvarname,legend = levels(cut(colourvar,breaks = n.colours)),col = pal(n.colours), pch = symbol) #do not change
```
