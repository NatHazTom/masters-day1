Now that we have learnt to fit linear and nonlinear regressions, the aim is to fit and compare models to estimate pressure using cpx chemistry.

Choose an independent variable (X-variable) to fit to Pressure. Then fit a statistical model and examine how well the model fits the data. Find which chemical elements and which type of relationship (linear vs. nonlinear) best fit the data. You can compare the results using the R<sup>2</sup> value and the graph plotted below with the fitted line.

```r
df <- read.table("CPX_experiments.txt",sep="\t",header=TRUE)
```

```r
Y <- df$P_GPa
X <- df$SiO2.cpx
ylab <- "Pressure (GPa)"
xlab <- "SiO2 (wt. %)"

plot(X,Y,xlab=xlab,ylab=ylab)
```

```r
model.data <- data.frame(Y,X)
linearMod <- lm(Y ~ X, data=model.data)
summary(linearMod)

plot(X,Y,pch=21,col="grey",bg="black",xlab=xlab,ylab=ylab)
lines(sort(X), fitted(linearMod)[order(X)], col='red', type='l',lwd=2) 
```
